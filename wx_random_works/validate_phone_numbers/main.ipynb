```python
import itertools
import sys
# https://geekflare.com/phone-number-validation-api/
# https://www.neutrinoapi.com/api/sdks/
from neutrino_api.neutrino_api_client import NeutrinoAPIClient
```

```python
def get_phone_possibilities(phone_number: str):
    missing_digits = sum([1 for x in phone_number if x=="*"])
    phone_number = phone_number.replace("*", "{}")
    missing_combinations = itertools.combinations_with_replacement(list(range(10)), missing_digits)
    for missing_combination in missing_combinations:
        yield phone_number.format(*missing_combination)

phone = "+628********3"
phone_possibilities = list(get_phone_possibilities(phone))
len(phone_possibilities)
```

```python
client = NeutrinoAPIClient("<USER_ID>", "<API_KEY>")

params = {"number": "+628000000003", "country_code": "", "ip": ""}
response = client.phone_validate(params)

if response.is_ok():
    data = response.data

    response_data = {"international_number": data.get("international-number"),
                     "international_calling_code": data.get("international-calling-code"),
                     "prefix_network": data.get("prefix-network"),
                     "is_valid": data.get("valid"),
                     "is_mobile": data.get("is-mobile"),
                     "type": data.get("type"),
                     "country": data.get("country"),
                     "country_code": data.get("country-code"),
                     "location": data.get("location")}

else:
    print("API Error: {0}, Error Code: {1}, HTTP Status Code: {2}".format(response.error_message, response.error_code, response.status_code), file = sys.stderr)
    if response.error_cause:
        print("Error Caused By: {0}".format(response.error_cause), file = sys.stderr)
```
